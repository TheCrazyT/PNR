```python
import matplotlib.pyplot as plt
from matplotlib.patches import Wedge
from matplotlib.text import Text
import math
from numpy import arange

def addWedge(axes, mod, cnt, r, start=1, color="r"):
  step = 360.0 / (cnt - start + 1)
  k = start-1
  for i in arange(0.0,360.0,step):
    col = color
    k += 1
    if((k % mod) == 0):
      col = "w"
    fov = Wedge(center = (0.5,0.5),
                r = r,
                theta1 = 180.0 - (i+step),
                theta2 = 180.0 - i,
                facecolor = col,
                edgecolor = "black",
                alpha = 1,
                linewidth = 1,
                linestyle = 'dashed',
                antialiased = True
               )
    axes.add_artist(fov)

def addTexts(axes, cnt, cntTot, start=1, txtStep = 0.01):
  step = 360 / (cnt - start + 1)
  end = cntTot + 1
  #print(start)
  #print(end)
  for i in range(start,end):
    i2 = i - start
    r = (0.3 + txtStep * i2)
    a = (i2 * step) * math.pi / 180.0 - 90.0 * math.pi / 180.0 + step / 2 * math.pi / 180.0
    x = 0.5 + math.sin(a) * r
    y = 0.5 + math.cos(a) * r
    text = "%d" % i
    axes.text(x,y,text)
```

```python
figure, axes = plt.subplots(1,1)
cntMax = 2*3
addWedge(axes = axes,
         mod = 2,
         cnt = cntMax,
         r = 0.25)
addWedge(axes = axes,
         mod = 3,
         cnt = cntMax,
         r = 0.25-0.05,
         color = "g")
addTexts(axes = axes,
         cnt = cntMax,
         cntTot = cntMax)
axes = plt.gca()
axes.set_aspect('equal')
figure.patch.set_facecolor('white')
plt.axis('off')
plt.show()
```

```python
figure, axes = plt.subplots(1,1, figsize=(15, 15))
cntMax = 7**2-1
start = 5**2
addWedge(start = start,
         axes = axes,
         mod = 2,
         cnt = cntMax,
         r = 0.25)
addWedge(start = start,
         axes = axes,
         mod = 3,
         cnt = cntMax,
         r = 0.25-0.05*1,
         color = "g")
addWedge(start = start,
         axes = axes,
         mod = 5,
         cnt = cntMax,
         r = 0.25-0.05*2,
         color = "b")
addTexts(start = start,
         axes = axes,
         cnt = cntMax,
         cntTot = cntMax,
         txtStep = 0.001)
axes = plt.gca()
axes.set_aspect('equal')
figure.patch.set_facecolor('white')
plt.axis('off')
plt.show()
```
